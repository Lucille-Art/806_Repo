## Challenge 1

#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1

%matplotlib inline

plt.plot(x,y, label='linear')
plt.plot(x,z, label='quadratic',color='green')
plt.legend()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2

fig, axes=plt.subplots(1,2)

axes[0].plot(x,y, label='linear')

axes[1].plot(x,z, label='quadratic', color='green')

axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

axes[1].set_xlabel('x label')
axes[1].set_ylabel('y label')

axes[0].set_title('Linear')
axes[1].set_title('Quadratic')



plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here

fig, axes=plt.subplots(1,2)

plt.figure(figsize=(20,20))

axes[0].plot(x,y, label='linear')

axes[1].plot(x,z, label='quadratic', color='green')

axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

axes[1].set_xlabel('x label')
axes[1].set_ylabel('y label')

axes[0].set_title('Linear')
axes[1].set_title('Quadratic')



plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

x = np.arange(0,100)
y1=x*2
y2=np.exp(x)


fig, axes=plt.subplots(1,2)

axes[0].plot(x,y1, label='quadractic')

axes[1].plot(x,y2, label='exponential', color='green')

axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

axes[1].set_xlabel('x label')
axes[1].set_ylabel('y label')

axes[0].set_title('Quadratic')
axes[1].set_title('Exponential')


plt.yscale("log")
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here

x = np.arange(0,100)
y1=x*2
y2=np.exp(x)


fig, axes=plt.subplots(1,2)
plt.legend()
axes[0].plot(x,y1, label='quadractic')
plt.legend()
axes[1].plot(x,y2, label='exponential', color='green')

axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

axes[1].set_xlabel('x label')
axes[1].set_ylabel('y label')

axes[0].set_title('Quadratic')
axes[1].set_title('Exponential')


plt.yscale("log")
plt.legend()
```
